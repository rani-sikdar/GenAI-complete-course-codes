```python
file_ ='/content/spam_classifier_dataset.csv'
```

```python
import pandas as pd

messages_df = pd.read_csv(file_)
messages_df
```

```python
messages_df['message']
```

```python
import nltk
import re

nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('wordnet')
```

```python
corpus =[]
for i in range(len(messages_df)):
  sentences =re.sub('[^a-zA-z]',' ',messages_df['message'][i])
  sentences = sentences.lower().split()

  sentences = [lemmatizer.lemmatize(word)  for word in sentences if word not in set(stopwords.words('english'))]
  sentences = ' '.join(sentences)
  corpus.append(sentences)
```

```python
corpus
```

```python
# implementing TF-IDF

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=100)
```

```python
X = tfidf.fit_transform(corpus).toarray()
```
